**Handling the missing values**

- Cabin: It contains many missing values making it unreliable for analysis, so we drop it from further consideration.
- Age: The missing values are filled using the median age.

```python
df.drop("Cabin", axis=1, inplace=True)
df["Age"].fillna(df["Age"].median(), inplace=True)
df["Embarked"].fillna(df["Embarked"].mode()[0], inplace=True)
df.isnull().sum()
```

#### **Outlier Detection and Handling**






```python
df.nunique()
```

```python
numerical_columns = ["Age", "Fare", "Parch", "SibSp"]
categorical_columns = ["Survived", "Pclass", "Sex", "Embarked"]

plt.figure(figsize=(15, 10))
for i, col in enumerate(numerical_columns, 1):
    plt.subplot(2, (len(numerical_columns) + 1) // 2, i)
    sns.boxplot(x=df[col], color='c')
    plt.title(f"{col} Boxplot")
    plt.xlabel(col, fontsize=10)

plt.tight_layout()
plt.show()
```

- The box plots show the pressence of outliers in numerical columns.

**Handling outliers:**


```python
df = df[df['Parch'] <= 8]
df = df[df['SibSp'] <= 6]
df.reset_index(drop=True, inplace=True)
df
```

```python
print(f"skewness of 'Fare' {df['Fare'].skew()}")
print(f"skewness of 'Age' {df['Age'].skew()}")
```

- Fare: The skewness of is large we have to use the IQR method.
- Age: The skewness of is less, so the data is almost normally distributed we will use Z-score method.

- **IQR Method** 

```python
Q1 = df["Fare"].quantile(0.25)
Q3 = df["Fare"].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
print(f"Number of outliers for Fare with IQR method {df[(df["Fare"] < lower_bound) | (df["Fare"] > upper_bound)].shape[0]}")
df = df[(df["Fare"] >= lower_bound) & (df["Fare"] <= upper_bound)]
df.reset_index(drop=True, inplace=True)
```

- **Z score method**

```python
from scipy import stats

df["Z_Score"] = abs(stats.zscore(df["Age"]))
print(f"Number of outliers in 'Age' with Z-score method: {df[df["Z_Score"] >= 3].shape[0]}")
df = df[df["Z_Score"] < 3]
df.drop(columns=["Z_Score"], inplace=True)
df.reset_index(drop=True, inplace=True)
df
```

### **Standardize the categorical value**

**Drop of redundant and irrelevant data**
- Name and PassengerID: These columns are irrelevant for our analysis so, drop them.

- Ticket: Redundant column. Ticket classes like first-class vs. third-class are already represented by the Pclass 
  column. Expensive ticket numbers (often associated with first-class) are indirectly captured by the Fare column.


**Encoding**

- Sex: Encoded as 0 (female) and 1 (male) fornumerical analysis.
- Embarked: Encoded as 0 (C) and 1 (Q) and 2(S) for numerical analysis.

```python
df.drop(columns=["Name", "Ticket", "PassengerId"], inplace=True)
df['Sex'] = df['Sex'].str.lower().str.strip()
df['Embarked'] = df['Embarked'].str.upper().str.strip()

df_encoded = df.copy()
df_encoded["Sex"] = df_encoded["Sex"].map({"male": 0, "female": 1})
df_encoded["Embarked"] = df_encoded["Embarked"].map({"C": 0, "Q": 1, "S": 2})
df_encoded
```

- Third-class passengers had lower survival rates, further supporting the link between social status and survival probability.  

### 6. Survived vs. Fare (Positive Correlation)  
- Higher fares were associated with higher survival rates, confirming that wealthier passengers had a better chance of survival.  

### 7. Survived vs. Age (Weak Negative Correlation)  
- Older passengers had a slightly lower survival rate, but the correlation is not very strong.  
- This suggests that age alone was not a major determinant of survival but might have played a role in combination with other factors like class and sex.
